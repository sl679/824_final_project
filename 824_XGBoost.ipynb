```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')
```

```python
#read in dataset
df=pd.read_csv('data.csv')
```

```python
#take a look at the dataframe
df
```

```python
#check data discription
df.info()
```

```python
#since we have non-null we skip the step to check for null
df.isna().sum()
```

```python
df.drop(['Unnamed: 32'],axis =1,inplace = True)
```

```python
#M = malignant, B = benign
#M: Malignant tumors are cancerous
#B: Benign tumors are noncancerous
sns.countplot(df.diagnosis)
#Set B as 0, M as 1
df.diagnosis.replace(to_replace=['B','M'],value=[0,1],inplace=True)
```

```python
#plot the corellation matrix
corr = df.corr()
plt.figure(figsize=(20,20))
sns.heatmap(corr, cmap='cividis', annot = True)
plt.show()
```

```python
#acc=df[df.columns[:]].corr()['diagnosis']
#print('Each feature and its correlations is: \n',acc)
```

```python
#select important features
#drop highly correlated features
df_re = df[['diagnosis', 'radius_mean', 'concavity_mean', 'compactness_worst', 
         'smoothness_mean', 'symmetry_mean','area_se','fractal_dimension_se',
         'texture_worst', 'smoothness_worst', 'symmetry_worst',
       'fractal_dimension_worst']]
df_re.head()
```

```python
#plot the corellation matrix of the reduced df
corr = df_re.corr()
plt.figure(figsize=(18,18))
sns.heatmap(corr, cmap='cividis', annot = True)
plt.show()
```

```python
#drop, scale, split dataset
x= df.drop(['diagnosis','id'],axis =1)
y = df.diagnosis
sc = StandardScaler()
x = sc.fit_transform(x)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
#logistics regression and accuracy
acc = {}
lr = LogisticRegression()
lr.fit(x_train,y_train)
acc['Logistic Regression'] = accuracy_score(y_test,lr.predict(x_test))
accuracy_score(y_test,lr.predict(x_test))
```

```python
# ada boost and accuracy
abc = AdaBoostClassifier()
abc.fit(x_train,y_train)
acc['Ada Boost'] = accuracy_score(y_test,abc.predict(x_test))
accuracy_score(y_test,abc.predict(x_test))
```

```python
# XGboost and accuracy
xgb = XGBClassifier()
xgb.fit(x_train,y_train)
acc['XGBoost'] = accuracy_score(y_test,xgb.predict(x_test))
accuracy_score(y_test,xgb.predict(x_test))
```

```python
# create accuracy comparison plot of each method
key = list(acc.keys())
val = [float(acc[k]) for k in key]
col = sns.color_palette("YlGnBu_r",8)
col = col.as_hex()
fig,acc = plt.subplots(figsize = (16,6))
acc = sns.barplot(val,key,palette="YlGnBu")
acc.set_xlim(0,1.1)
for i in range(0,len(key)):
    acc.text(val[i]-0.05,i,str(np.round(val[i],4)),fontdict = dict(color = col[i],fontsize = 18,ha = 'center'),weight = 'bold')
```

```python
# now do the same steps but with the feature-reduced dataset

#drop, scale, split dataset
x= df_re.drop(['diagnosis'],axis =1)
y = df_re.diagnosis
sc = StandardScaler()
x = sc.fit_transform(x)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=28)

#logistics regression and accuracy
acc = {}
lr = LogisticRegression()
lr.fit(x_train,y_train)
acc['Logistic Regression'] = accuracy_score(y_test,lr.predict(x_test))
accuracy_score(y_test,lr.predict(x_test))

# ada boost and accuracy
abc = AdaBoostClassifier()
abc.fit(x_train,y_train)
acc['Ada Boost'] = accuracy_score(y_test,abc.predict(x_test))
accuracy_score(y_test,abc.predict(x_test))

# XGboost and accuracy
xgb = XGBClassifier()
xgb.fit(x_train,y_train)
acc['XGBoost'] = accuracy_score(y_test,xgb.predict(x_test))
accuracy_score(y_test,xgb.predict(x_test))

# create accuracy comparison plot of each method
key = list(acc.keys())
val = [float(acc[k]) for k in key]
col = sns.color_palette("YlGnBu_r",8)
col = col.as_hex()
fig,acc = plt.subplots(figsize = (16,6))
acc = sns.barplot(val,key,palette="YlGnBu")
acc.set_xlim(0,1.1)
for i in range(0,len(key)):
    acc.text(val[i]-0.05,i,str(np.round(val[i],4)),fontdict = dict(color = col[i],fontsize = 18,ha = 'center'),weight = 'bold')
```
